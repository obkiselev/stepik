```python
'''Скачайте файл. В нём указан адрес другого файла, который нужно скачать с 
использованием модуля requests и посчитать число строк в нём.'''
import requests
with open('file_in_out/dataset_3378_2.txt', 'r', encoding='utf-8') as inf:
    url = inf.readline().strip()
    print(url)

r = requests.get(url)

ouf = open('file_in_out/dataset_3378_2_out.txt', 'w')
ouf.write(r.text)
ouf.close()

print(len(r.text.splitlines()))
print(' ')
print(r.text)
```
